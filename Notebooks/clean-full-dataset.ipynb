## Removing duplicates (URLs with same host)

```python
%pylab inline
%matplotlib widget

import sys 
import pandas as pd

sys.path.append('..')
import bad_URLs_API
```

```python
URL_REGEX = ("^((?P<scheme>[^:/?#]+):(?=//))?(//)?(((?P<login>[^:]+)" + 
    "(?::(?P<password>[^@]+)?)?@)?(?P<host>[^@/?#:]*)(?::(?P<port>\d+)?)?)?" + 
    "(?P<path>[^?#]*)(\?(?P<query>[^#]*))?(#(?P<fragment>.*))?")
```

```python
combined_urls = pd.read_csv("../Data/recomposed_combined_urls_backup.csv", index_col=0)
combined_urls.head()
```

```python
combined_urls.url = combined_urls.url.str.replace("@", "a")
combined_urls.url = combined_urls.url.str.replace("///", "//")
```

```python
url_parse_groups = ["scheme", "login", "password", "host", "port", "path", "query", "fragment"]
url_parse_res = pd.concat(
    [combined_urls, combined_urls.url.str.extract(URL_REGEX)[url_parse_groups]],
    axis=1
)    
print(url_parse_res.shape)
url_parse_res.head()
```

```python
url_parse_res["path_len"] = url_parse_res.path.str.len()
url_parse_res.sort_values(["host", "path_len"], inplace=True)
url_parse_res.reset_index(drop=True, inplace=True)
url_parse_res.head(20)
```

```python
url_parse_res.drop_duplicates(["host"], inplace=True, keep="last", ignore_index=True)
url_parse_res.dropna(subset=["url", "host"], inplace=True)
```
